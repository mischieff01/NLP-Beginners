# Tokenization

```python
import pandas as pd
import PyPDF2
```

```python
from PyPDF2 import PdfReader
```

```python
# load the pdf file
pdf_file = PdfReader("machine_learning_detailed.pdf")
```

```python
# extract text from all file
text = ""
for page in pdf_file.pages:
    text += page.extract_text()

print(text)
```

```python
# print page 1 of file
page = pdf_file.pages[0]
page_text = page.extract_text()
print(page_text)
```

#### Brute force approach 

```python
# lets tokenize the page 1 
tokens = page_text.split()
```

```python
tokens
```

#### Now we use the nltk library for effecient tokenization

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
# tokenize into sentences
sentences = sent_tokenize(page_text)
```

```python
sentences
```

```python
# Word tokenize 
words = word_tokenize(page_text)
words
```
